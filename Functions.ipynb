```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5 , 5 , 100)
y = x**2
plt.plot(x,y)
```

```python
x = np.linspace(-10 , 10 , 100)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = np.linspace(-10 , 10 , 100)
y = np.sin(2*x)
plt.plot(x,y)
```

```python
x = np.linspace(-10 , 10 , 100)
y = np.tanh(x)
plt.plot(x,y)
```

```python
x = np.linspace(-10 , 10 , 100)
y = np.maximum(0 , x)
plt.plot(x, y)
```

```python
x = np.linspace(-10 , 10 , 100)
y = 1/(1+np.exp(-x))
plt.plot(x,y)
```
